## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Make a copy of dataset so all manipulation and aggregation don't affect the original dataset
df  = df_hba.copy()
# Data Cleaning:
# to remove duplicate values we use function drop_duplicate to delete duplicate rows
df.drop_duplicates(inplace=True)
# reset index
df.reset_index()
````

````python
# Here we check the null/missing values
df.isnull().sum().sort_values(ascending=False)
````

````python
# It is clearly shows that company's column have maximun null values that is approx 93% 
company_null_value_percentage = df["company"].isnull().sum()/df.shape[0]*100
print("Percentage of company's column null value :" ,company_null_value_percentage) 
````

````python
# Its better to drop the company column because the number of missing values is extremely high
df.drop(['company'],axis = 1,inplace = True)
````

````python
# Checking the percentage of null value for agent column
agent_null_value_percentage = df["agent"].isnull().sum()/df.shape[0]*100
print("Percentage of agent's column null value : ",agent_null_value_percentage)
````

````python
# As we seen there is minimal null values in agent column , let's fill these value by 0
df['agent'].fillna(0,inplace = True)
# Re- Check that column has no null value
df['agent'].isnull().sum()
````

````python
# Now, Checking the percentage of null value for country column
country_null_value_percentage = df["country"].isnull().sum()/df.shape[0]*100
print("Percentage of country's column null value :",country_null_value_percentage)
````

````python
# As we see there is less null values in country column , let's fill it by "other" as country name
df['country'].fillna('other',inplace = True)
# Re- Check that column has no null value
df['country'].isnull().sum()
````

````python
# Now, Checking the percentage of null value for children column
children_null_value_percentage = df["children"].isnull().sum()/df.shape[0]*100
print('children_null_value_percentage: ',children_null_value_percentage)   
````

````python
# As we see there is less null values in children column , let's fill it by 0 
df['children'].fillna(0,inplace = True)
# Re- Check that column has no null value
df['children'].isnull().sum()
````

````python
# Check whether dataset has any other null value
df.isnull().sum()      # As we seen some column has a null value 
````

````python
# Data Manipulation:
# To do data manipulation firstly look on info of dataset
df.info()
````

````python
# As we seen that 'children' and 'agent' column have float data type which is not correct because 'children' column has the number of children which should be of int type
# and 'agent' column has the id of agent which should also be of int type. So lets change the datatype of both column
df[['children','agent']]  = df[['children','agent']].astype(int)
````

````python
# As we seen "is_canceled" and "is_repeated_guest" column have two values i.e. (0) for not canceled or repeated or (1) for canceled or repeated 
# its not seems to be reasonable So, lets change it
df['is_canceled'] = df['is_canceled'].replace([0,1],['not canceled','canceled'])
df['is_repeated_guest'] = df['is_repeated_guest'].replace([0,1],['not repeated guest','repeated guest'])
````

````python
# To get total guest we just simply add adult, children and babies columns and make a new column name as total guest at index 13
df.insert(13, 'total_guest', (df['adults']+df['children']+df['babies']), True)
````

````python
#To get information of total stay nights by guests we add a new column at index 10 named "total_stays_night" that contains the add of totalstays_in_weekend_nights and week night
df.insert(10, 'total_nights', (df['stays_in_weekend_nights'] + df['stays_in_week_nights']), True)
````

#### Chart - 2

````python
# Chart - 2 visualization code
# To find the total bookings we group the hotel column by is_canceled column 
total_bookings = df.groupby('hotel')['is_canceled'].value_counts()
# We use loc method to select the particular values of the column
city_hotel_not_canceled = total_bookings.loc['City Hotel', 'not canceled']
resort_hotel_not_canceled = total_bookings.loc['Resort Hotel', 'not canceled']
# print the result
print("Total bookings not canceled for City Hotel:", city_hotel_not_canceled)
print("Total bookings not canceled for Resort Hotel:", resort_hotel_not_canceled)
````

````python
# Total bookings not canceled for each hotel
labels = ['City Hotel', 'Resort Hotel']
sizes = [city_hotel_not_canceled, resort_hotel_not_canceled]

# Pie chart properties
colors = ['#ff9999', '#66b3ff']
explode = (0.1, 0)  # Explode the first slice (City Hotel)

# Create the pie chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',
        shadow=True, startangle=90)

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.title('Total Bookings Not Canceled by Hotel',bbox={'facecolor':'0.8','pad':1})

plt.legend(title = 'Hotels' ,loc = "upper right", prop={'size': 8})

# Display the pie chart
plt.show()
````
